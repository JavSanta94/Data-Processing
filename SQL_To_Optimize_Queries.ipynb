```python
import kagglehub
import pandas as pd
import sqlite3

path = kagglehub.dataset_download("die9origephit/amazon-data-science-books")
print("Path to dataset files:", path)

# Aquí debes buscar la ruta completa al archivo CSV dentro del directorio descargado
csv_file_path = path + '/final_book_dataset_kaggle2.csv'

# Leer el archivo CSV con pandas
df = pd.read_csv(csv_file_path)

conn = sqlite3.connect(':memory:')

# Guardar el DataFrame en una tabla SQLite
df.to_sql('amazon_books', conn, index=False, if_exists='replace')

# Mostrar las primeras filas del dataset
print(df.head())
```

# Queries Basic
Esta consulta busca todos los libros escritos por el autor "[ Dr Dhaval Maheta]" en la tabla ``amazon_books`` y devuelve las columnas ``title``, ``price``, ``pages``, ``publisher``, y ``author`` para esos libros.

```python
query1 = """
SELECT title, price, pages, publisher, author
FROM amazon_books
WHERE author = "[ Dr Dhaval Maheta]"
"""
result1 = pd.read_sql_query(query1, conn)

print(result1)
```

Esta consulta busca todos los libros que tienen un **precio menor a 10** y **menos de 200 páginas** en la tabla ``amazon_books`` y devuelve las columnas ``title``, ``price``, ``pages``, ``publisher``, y ``author`` para esos libros.

```python
query2 = """
SELECT title, price, pages, publisher, author
FROM amazon_books
WHERE price < 10
AND pages < 200
"""
result2 = pd.read_sql_query(query2, conn)
print(result2)
```

Esta consulta selecciona los libros que cumplen con las siguientes condiciones:

1. **El autor no es igual a** ``NaN``.
2. **El precio es menor a 10 o el número de páginas es menor a 200.**

Este es un ejemplo de cómo las condiciones **AND** y **OR** pueden combinarse para realizar consultas más flexibles y complejas.

```python
query3 = """
SELECT title, price, pages, publisher, author
FROM amazon_books
WHERE NOT author = "NaN"
AND price < 10
OR pages < 200
"""
result3 = pd.read_sql_query(query3, conn)
print(result3)
```

Esta consulta selecciona todos los libros de la tabla ``amazon_books`` que han sido escritos por **"[ Dr Dhaval Maheta]"** o **"[ Chad Knowles]"**, y devuelve las columnas ``title``, ``price``, ``pages``, ``publisher``, y ``author`` para esos libros.

Es una forma sencilla y eficiente de obtener los libros de uno o varios autores específicos usando el operador **IN**.

```python
query4 = """
SELECT title, price, pages, publisher, author
FROM amazon_books
WHERE author IN ("[ Dr Dhaval Maheta]","[ Chad Knowles]")
"""
result4 = pd.read_sql_query(query4, conn)
print(result4)
```

Esta consulta selecciona todos los libros de la tabla ``amazon_books`` cuyo *precio* se encuentra entre *10 y 13 unidades de la moneda* (inclusive), y devuelve las columnas ``title``, ``price``, ``pages``, ``publisher``, y ``author`` para esos libros.

El operador **BETWEEN** es útil cuando deseas trabajar con rangos de valores y lo hace más legible que usar múltiples condiciones con **AND**.

```python
query5 = """
SELECT title, price, pages, publisher, author
FROM amazon_books
WHERE price BETWEEN 10 AND 13
"""
result5 = pd.read_sql_query(query5, conn)
print(result5)
```

Esta consulta selecciona todos los libros de la tabla ``amazon_books`` cuyo editor (publisher) comienza con la palabra **"Independently"** y devuelve las columnas ``title``, ``price``, ``pages``, ``publisher``, y ``author`` para esos libros.

El uso de **LIKE con %** es útil para realizar búsquedas parciales o patrones en texto, como encontrar todos los libros de un editor que comience con un prefijo específico.

```python
query6 = """
SELECT title, price, pages, publisher, author
FROM amazon_books
WHERE publisher LIKE "Independently%"
"""
result6 = pd.read_sql_query(query6, conn)
print(result6)
```

# Subquery Basic
Esta consulta selecciona todos los libros cuyo precio es superior al precio promedio de todos los libros en la base de datos. Esto se logra mediante una subconsulta que calcula el promedio de los precios (**AVG(price)**), y luego se filtran los libros cuyo precio es mayor que ese valor promedio.

Es un ejemplo de cómo usar subconsultas en SQL para realizar comparaciones con valores agregados (en este caso, el precio promedio de los libros).

```python
query7 = """
SELECT title, price, pages, publisher, author
FROM amazon_books
WHERE price > (SELECT AVG(price) FROM amazon_books)
"""
result7 = pd.read_sql_query(query7, conn)
print(result7)
```

Consulta una lista de libros de la tabla ``amazon_books`` con detalles sobre el título, precio, número de páginas, editor y autor de cada libro. Además, para cada libro, calcula el precio promedio de todos los libros disponibles en la base de datos. El resultado se muestra en una tabla con los datos solicitados y el precio promedio calculado para todos los libros.

```python
query8 = """
SELECT title, price, pages, publisher, author, (SELECT AVG(price) FROM amazon_books) AS Promedio_Precio
FROM amazon_books
"""
result8 = pd.read_sql_query(query8, conn)
print(result8)
```

# Funciones SQL
Este código obtiene información sobre los libros en la tabla ``amazon_books``, incluyendo el título, precio, número de páginas, editorial y autor. Luego, redondea el precio de cada libro a un decimal y lo presenta en la columna ``price_round``. Al final, se imprime el resultado con todos los detalles, incluyendo el precio redondeado.

```python
query9 = """
SELECT title, price, pages, publisher, author, ROUND(price,1) AS price_round
FROM amazon_books
"""
result9 = pd.read_sql_query(query9, conn)
print(result9)
```

Existen otras funciones que se pueden agregar al select para columnas númericas como:

- TRUNCATE (elimina decimales)
- CEILING (redondear para arriba)
- FLOOR (redondear para abajo)
- ABS (muestra el valor absoluto)
- RAND (crea valores decimales aleatorios)

Tambien hay algunas que se pueden usar con columnas tipo **string**:

- LENGTH (muestra la cantidad de caracteres incluye espacios)
- UPPER (convierte el texto a mayúscula)
- LOWER (convierte el texto a minúscula)
- LTRIM/RTRIM (borra todos los espacios a la izquierda/derecha de un texto)
- LEFT (Extrae x cantidad de caracteres de una palabra por la izquierda y luego los muestra)
- RIGHT Extrae x cantidad de caracteres de una palabra por la derecha y luego los muestra)
- TRIM (quita todos los espacios de más al principio y al final)
- SUBSTRING (quita caracteres fuera del rango indicado y muestra el resultado)
- LOCATE (muestra la posición de una letra en una palabra)
- REPLACE (reemplaza un caracter o una cadena de caracteres buscando por uno nuevo)
- CONCAT (une dos cadenas)

```python
query10 = """
SELECT title, price, pages, publisher, UPPER(author)
FROM amazon_books
"""
result10 = pd.read_sql_query(query10, conn)
print(result10)
```

las funciones **SUBSTR, LENGTH, y INSTR** para manejar la extracción del año de la cadena de texto en tu columna. Dependiendo de cómo está estructurada la cadena (si el año está al final o dentro de paréntesis).

```python
query10_2 = """
SELECT title, price, pages, publisher, SUBSTR(publisher, LENGTH(publisher) - 5 + 1, 4) AS year_publisher
FROM amazon_books
LIMIT 10
"""
result10_2 = pd.read_sql_query(query10_2, conn)
print(result10_2)
```

# Agregaciones
Este código obtiene, para cada autor en la tabla ``amazon_books``, el precio total de todos sus libros (suma de los precios) y el número de libros que ha publicado (conteo de los títulos). Los resultados se agrupan por autor. Al final, el código imprime una tabla que muestra estos tres valores para cada autor: ``author``, ``precio_total`` y ``numero_libros``.

```python
query11 = """
SELECT author, SUM(price) AS precio_total, COUNT(title) AS numero_libros
FROM amazon_books
GROUP BY 1
"""
result11 = pd.read_sql_query(query11, conn)
print(result11)
```

Este código obtiene, para cada autor en la tabla ``amazon_books``, el precio total de sus libros y el número de libros que ha publicado. Luego, solo muestra a aquellos autores que tienen más de 2 libros en la base de datos. Los resultados se agrupan por autor, y se calcula el precio total y el número de libros de cada uno. El filtro ``HAVING numero_libros > 2`` asegura que solo se muestren los autores con más de dos libros. Finalmente, el código imprime los resultados en forma de tabla.

```python
query12 = """
SELECT author, SUM(price) AS precio_total, COUNT(title) AS numero_libros
FROM amazon_books
GROUP BY 1
HAVING numero_libros > 2
"""
result12 = pd.read_sql_query(query12, conn)
print(result12)
```

### ¿Cuáles son los 10 productos con mayor demanda total?

```python
query14 = """
SELECT p.subcategoria, SUM(d.demanda) AS total_demanda
FROM demanda d
LEFT JOIN catalogo_productos p ON d.id_producto = p.id_producto
GROUP BY p.categoria
ORDER BY total_demanda DESC
LIMIT 10
"""
result14 = pd.read_sql_query(query14, conn)
print(result14)
```

### ¿Cuál es la demanda total por categoría y subcategoría de producto?

```python
query15 = """
SELECT p.categoria, p.subcategoria, SUM(d.demanda) AS total_demanda
FROM demanda d
JOIN catalogo_productos p ON d.id_producto = p.id_producto
GROUP BY p.categoria, p.subcategoria
ORDER BY total_demanda DESC
"""
result15 = pd.read_sql_query(query15, conn)
print(result15)
```

### ¿Cuál es la demanda promedio mensual por producto en los últimos 6 meses?

```python
query16 = """
SELECT p.subcategoria, AVG(d.demanda) AS demanda_promedio
FROM demanda d
JOIN catalogo_productos p ON d.id_producto = p.id_producto
JOIN (
    SELECT MAX(d.date) AS max_date
    FROM demanda d
) AS fecha_maxima
WHERE d.date >= DATE(fecha_maxima.max_date, '-6 months')
GROUP BY p.subcategoria
ORDER BY demanda_promedio DESC
"""
result16 = pd.read_sql_query(query16, conn)
print(result16)
```

### ¿Cuáles son los productos más vendidos durante temporadas estacionales?

```python
query17 = """
SELECT p.subcategoria, p.categoria, SUM(d.demanda) AS demanda_estacional
FROM demanda d
JOIN catalogo_productos p ON d.id_producto = p.id_producto
WHERE p.estacional = '1'
GROUP BY p.subcategoria, p.categoria
ORDER BY demanda_estacional DESC
"""
result17 = pd.read_sql_query(query17, conn)
print(result17)
```

### ¿Qué productos premium tienen una demanda superior a 1000 unidades?

### ¿Cuál es la tendencia de demanda del producto bebidas en los últimos 3 meses?

```python
query20 = """
SELECT p.subcategoria, strftime('%Y-%m', d.date) AS mes, SUM(d.demanda) AS demanda_mensual
FROM demanda d
JOIN catalogo_productos p ON d.id_producto = p.id_producto
JOIN (
    SELECT MAX(d.date) AS max_date
    FROM demanda d
) AS fecha_maxima
WHERE p.categoria = 'bebidas' AND d.date >= DATE(fecha_maxima.max_date, '-3 months')
GROUP BY p.subcategoria, mes
ORDER BY mes
"""
result20 = pd.read_sql_query(query20, conn)
print(result20)
```

### ¿Cuál es la correlación entre el tamaño del producto y la demanda?

```python
query21 = """
SELECT p.tamaño, SUM(d.demanda) AS total_demanda, AVG(d.demanda) AS Promedio_demanda
FROM demanda d
JOIN catalogo_productos p ON d.id_producto = p.id_producto
GROUP BY p.tamaño
ORDER BY total_demanda DESC
"""
result21 = pd.read_sql_query(query21, conn)
print(result21)
```

### ¿Cuál es la distribución de la demanda por categorías en (2021-03) específico?

```python
query22 = """
SELECT p.categoria, SUM(d.demanda) AS demanda_mes
FROM demanda d
JOIN catalogo_productos p ON d.id_producto = p.id_producto
WHERE strftime('%Y-%m', d.date) = '2021-03'
GROUP BY p.categoria
ORDER BY demanda_mes DESC
"""
result22 = pd.read_sql_query(query22, conn)
print(result22)
```

### demanda total por categoría de producto en los últimos 6 meses

```python
query23 = """
WITH fecha_maxima AS (
    SELECT MAX(d.date) AS max_date
    FROM demanda d
)
SELECT p.categoria, SUM(d.demanda) AS demanda_total
FROM demanda d
JOIN catalogo_productos p ON d.id_producto = p.id_producto
JOIN fecha_maxima fm ON 1=1  -- Esto hace un "cross join" con la fecha máxima
WHERE d.date >= DATE(fm.max_date, '-6 months')
GROUP BY p.categoria
ORDER BY demanda_total DESC
"""
result23 = pd.read_sql_query(query23, conn)
print(result23)
```

### Obtener productos que tienen una demanda superior a la media de su categoría

```python
query24 = """
WITH demanda_categoria_promedio AS (
    SELECT p.subcategoria, AVG(d.demanda) AS demanda_promedio_categoria
    FROM demanda d
    JOIN catalogo_productos p ON d.id_producto = p.id_producto
    GROUP BY p.subcategoria
)
SELECT p.id_producto, p.subcategoria, d.demanda
FROM demanda d
JOIN catalogo_productos p ON d.id_producto = p.id_producto
JOIN demanda_categoria_promedio dc ON p.subcategoria = dc.subcategoria
WHERE d.demanda > dc.demanda_promedio_categoria
ORDER BY d.demanda DESC
"""
result24 = pd.read_sql_query(query24, conn)
print(result24)
```
