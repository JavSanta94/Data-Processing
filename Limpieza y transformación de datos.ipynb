# 2.1 Bucle for anidado
Como su nombre indica, significa un bucle for dentro de otro bucle for. Este concepto es muy avanzado, por lo que se tiene que tener bien claro el entendimiento del bucle for. A veces los bucles for anidados, nos ayudaran a solucionar problemas, que sin ellos no podríamos automatizarlos. Repasemos un momento el bucle for antes de pasar a ver el bucle for anidado. Ya sabemos que el bucle for nos permite recorrer una lista de valores. 

Ejemplo de bucle for, sobre un rango de valores con range():


````python
for i in range(5):
  print(i)
````

Ejemplo de bucle for en el que iteramos sobre las cinco primeras columnas del dataframe nba:

````python
for col in df.columns[:5]:
  print(col)
````

Acordaros siempre de la estructura básica de un bucle for, donde tenemos las palabras reservadas for e in, el iterador al cual le ponemos el nombre que queramos y sobre que estamos iterando. 

Pasemos a ver paso a paso cómo funciona un bucle for anidado y cómo nos puede ayudar en los problemas del mundo real. Un bucle for anidado no es más que un bucle dentro de otro bucle. Para entender el funcionamiento interno de un bucle for anidado, tenemos que verlo como si fueran las agujas de un reloj. Tenemos la aguja de las horas, que se mueve una vez a cada hora, pero por cada hora, la aguja de los minutos da la vuelta entera. Una vez realizada la vuelta de la aguja de los minutos, la aguja de las horas se mueve otra vez y se vuelve a repetir el mismo proceso con la aguja de los minutos. Esta sería la idea detrás del bucle for anidado, veámoslo más claro con un ejemplo. 

Crearemos dos listas, por las que iterará el primero y el segundo bucle, tendrán valores del 1 al 3 con range(1,4)


````python
primer_bucle = list(range(1,4))
print(primer_bucle)
segundo_bucle = list(range(1,4))
print(segundo_bucle)
````

Utilizaremos la función print(), para ver que sucede dentro de los bucles. Con ``print('--'*10)`` es simplemente un truco para añadir líneas y que quede mejor visualmente el resultado.

````python
for i in primer_bucle:

    print('--'*10)
    print('Primer bucle, iteración número:', i)
    print('--'*10)

    for j in segundo_bucle:
        print('Valor segundo bucle: ', j)
````

Primero, el primer bucle itera por su lista de valores, una vez pasa al primer valor de las lista que es el 1, pasamos a iterar sobre todos los valores en el segundo bucle. Una vez recorrido todos los valores del segundo bucle, pasamos de nuevo  al siguiente valor del primer bucle que será el 2. Así hasta haber completado todos los valores del primer bucle.  Como sucedía con las agujas del reloj, las horas serían los valores del primer bucle y los minutos los del segundo bucle.  No pasamos a  la siguiente hora, hasta que de la vuelta a todos los minutos.

Con la técnica de print(), vemos que sucede internamente en el bucle anidado y qué valores está iterando. Podemos ver claramente, cómo una vez iterado en el primer valor del primer bucle, iteramos sobre todo los valores del segundo bucle.

Acordaros que en cada iteración de un bucle for, ejecutamos todo el código que hay dentro de él, si en este caso tenemos otro  bucle for en su interior, a cada iteración del primer bucle, se ejecutará todo el bucle en el interior y así sucesivamente. Repite el proceso con ejemplos como el que acabamos de ver, para que interiorices y entiendas lo que pasa dentro del bucle for anidado.

Por ejemplo, vamos a utilizar el bucle for anidado, donde en el primer bucle iteramos sobre el valor de las 3 primeras columnas y el bucle anidado sobre los 3 primeros valores de cada columna.


````python
for col in df.columns[:3]:
  
  print('-'*15)
  print(col)
  print('-'*15)

  for i in range(3):
    print(df.loc[i, col])
````

Fijaros como en el primer bucle, obtenemos la primera columnas e imprimimos su nombre  y en el segundo bucle, obtenemos  los tres valores de la columna, una vez finalizado el segundo bucle, volvemos al primer for para pasar a la siguiente columna y nuevamente en el segundo for volvemos a iterar por los tres valores de la segunda columna.

 En la primera iteración, esto será lo que está pasando dentro del segundo for


````python
print(df.loc[0, 'Player'])
print(df.loc[1, 'Player'])
print(df.loc[2, 'Player'])
````

Donde la columna será la misa, pero el valor de las filas irá cambiando. Esta técnica de bucle for anidado, nos será muy útil en el módulo 3, cuando tengamos que automatizar la visualización de datos de eventing y tracking. ## 2.1.1. El problema de los duplicados. 

A veces en los datos nos podemos encontrar con duplicados, que no es más que una fila del dataframe que está repetida con exactamente los mismo valores. Para estos casos donde toda la fila tiene los mismo valores, podemos utilizar un función de pandas .drop_duplicates(). Esta función automáticamente eliminará las filas repetidas. Pero nosotros tenemos un problema un poco más complejo que ese. 

En el dataframe de las estadísticas de todos los jugadores de la NBA de toda una temporada, nos surge un problema a resolver y es que los jugadores que han jugado en varios equipos durante la temporada, los tendremos duplicados. Como por ejemplo, el caso de Trevor Ariza, que jugó en Portland y en Sacramento. Si filtramos el dataframe por su nombre, podremos verlo.


````python
df[df.Player == 'Trevor Ariza']
````

Este último, representa el total de las estadísticas que obtuvo el jugador, en los dos equipos. Para evitarnos problemas de duplicados, a los jugadores que les pase esto, nos quedaremos con los valores del  jugador donde Tm = 'TOT', eliminando la de los otros equipos. ¿Cómo hacemos esto? Ahora tenemos la variable names, que contiene una lista con el nombre de los jugadores duplicados. Pero lo que realmente nos interesa es saber los índices de las filas de estos jugadores. Para obtener los índices de los jugadores duplicados, filtraremos el dataset por su nombre y utilizaremos el atributo .index seguido .values, para obtener los índices de las filas en formato lista.  Por ejemplo con  'Trevor Ariza'  obtenemos  3 índices.

````python
df[df.Player =='Trevor Ariza'].index.values
````

## 2.1.2 Automatización

El dataframe utilizado, son datos extraídos de basketball-reference. Eso significa que siempre que tratemos con datos de la información de todos los jugadores siempre nos encontraremos con este problema. ¿Qué significa esto? Pues que tendremos que crear una automatización para quitarnos de encima este problema, ya que será un problema recurrente al utilizar estos datos.  

Vamos a encapsular todo el proceso del bucle for anidado dentro una función que llamaremos drop_duplicate_player(), a la que por parámetro le pasaremos el dataframe y la función nos retorna el dataframe sin jugadores duplicados.


````python
def drop_duplicate_player(df):
    
    # Nombres de los jugadores duplicados
    names = df[df.Tm =='TOT'].Player.values
    
    # Los índices con los que nos queremos quedar
    idx_good = df[df.Tm =='TOT'].index.values
    
    # Creamos lista vacía donde añadiremos los índices
    idx=[]

    # Iteramos sobre los nombre de la lista names
    for name in names:
        # Añadimos a la lista vacía los índices de las filas igual a ese nombre
        idx.append(df[df.Player == name].index.values)
        
    # Creamos la lista vacía donde irán los índices que queremos borrar
    drop_idx = []
    
    # El primer bucle itera sobre la lista de los índices duplicados 
    for i in idx:
        # El segundo bucle itera sobre los valores de cada lista del primer bucle
        for j in i:
            if j not in idx_good:
                drop_idx.append(j)
                
    df.drop(drop_idx, axis=0, inplace=True)
                
    return df
````

Ahora si importamos de nuevo los datos, los cuales tendrán el problema de los duplicados y le pasamos el dataframe a la función drop_duplicate_player() obtendremos automáticamente el dataframe sin jugadores duplicados.

````python
df = pd.read_csv('/content/drive/MyDrive/Mi curso/Data/clean_data/nba_19_pergame.csv')
df[(df.Player == 'Trevor Ariza') | (df.Player == 'Keita Bates-Diop')]
````

Vemos como nuevamente tenemos los jugadores duplicados. Pasemos a utilizar ahora nuestra función. 

````python
df = drop_duplicate_player(df)
df[(df.Player == 'Trevor Ariza') | (df.Player == 'Keita Bates-Diop')]
````

Gracias a nuestra función, ya no nos tendremos que preocupar más del problema de los jugadores en múltiples equipos.

# 2.2 List comprehension

¿Qué es esto de las list comprehension? No es nada más que una manera mejor y  más avanzada de realizar la técnica de la lista vacía en el bucle for. Ya hemos visto lo importante de crear una lista vacía en la cual después de un bucle for, vamos añadiendo los datos que necesitamos. Las list comprehension nos permiten realizar esta acción sin crear una lista vacía, sin utilizar .append() y ejecutando el bucle for en una sola línea de código. Veamos primero el ejemplo con el bucle for y después como podemos obtener el mismo resultado con las list comprehension.

Vamos a iterar sobre el nombre de los cinco primero jugadores del dataframe de la NBA 


````python
df.loc[:5, 'Player']
````

Como siempre, crearemos la lista vacía y dentro del bucle for utilizaremos .append() para añadir los valores a la lista vacía.

````python
names = []
for name in df.loc[:5, 'Player']:
  names.append(name)
````

Ahor si miramos dentro de names, tenemos la lista con los nombres 

````python
names
````

Vamos a ver como hacer todo este procedimiento con las list comprehension, en una sola línea de código.

````python
[name for name in df.loc[:5, 'Player']]
````

Hemos obtenido el mismo resultado sin tener que crear una lista vacía, ni utilizar .append(). La estructura de una list comprehension es siempre la misa:

Ejecutamos el bucle for dentro de los corchetes. 
Al principio siempre va el iterador que devolverá el bucle for. 
Por último el  bucle for de manera horizontal.


Lo que tenéis que entender es que antes de la palabra for, en la list comprehension va el iterador que en cada iteración tendrá un valor diferente y después del for, la estructura es la misma que cuando hacemos un bucle for convencional, el iterador y sobre que estamos iterando. 

Esta técnica es más eficiente que la de crear la lista vacía y utilizar .append(), ya que es más rápido y realizamos menos acciones. Al realizar el for dentro de la lista, automáticamente se está creando una lista sobre los valores que estamos iterando. Se que al principio puede costar un poco, pero su estructura siempre será la misma. A partir de ahora siempre que tengamos que usar la técnica de la lista vacía, utilizaremos las list comprehension, porque así lo hacen los profesionales y nosotros somos unos profesionales.

## 2.2.1. El condicional if dentro de la list comprehension

Hemos visto cómo utilizar los condicionales junto con el bucle for y cómo de esta manera nuestro código se hacía más poderoso. Con la list comprehension, también podemos usar los condicionales, empecemos viendo cómo usar el if.

Queremos simplemente seleccionar los valores que cumplan una condición. Acordaros, en el tema de limpieza de datos parte 2 del módulo 1, utilizamos el condicional if para encontrar las columnas que contenían el signo de porcentaje en los datos de InStat. Este fue el código que tuvimos que hacer para poder seleccionar dichas columnas. 

````python
jug_2b = pd.read_csv('/content/drive/MyDrive/Mi curso/Data/raw_data/jugadores_2b.csv')
````

````python
cols_pct = []

for col in jug_2b.columns:
  if col.__contains__('%'):
    cols_pct.append(col)
````

Donde al final, dentro de la variable cols_pct teníamos las columnas que estábamos buscando.

````python
cols_pct
````

Veamos ahora cómo sería con la list comprehension: 

````python
cols_pct = [col for col in jug_2b.columns if col.__contains__('%')]
cols_pct
````

Podemos ver como obtenemos exactamente el mismo resultado, añadiendo el condicional if después de la parte del bucle for.

Cuando en los datos de InStat, seleccionamos las columnas categóricas y utilizamos un bucle for y el condicional if, para añadir a una lista vacía las columnas numéricas que no se encontraban dentro de la lista de categóricas, lo hicimos de la siguiente manera:


````python
cols_cat = ['Nombre','Posición', 'Equipo','Pierna','Nacionalidad', 'Selección nacional']
cols_num = []

for col in jug_2b.columns:
  if col not in cols_cat:
    cols_num.append(col)

cols_num[:5]
````

Veamos como sería con la list comprehension:

````python
cols_num = [col for col in jug_2b.columns if col not in cols_cat]
cols_num[:5]
````

La estructura interna de una list comprehension siempre será la misma:

1. Abrimos corchetes. 
```python
[ ]
```


2. Colocamos el iterador, con el nombre que queramos darle.
```python
[col ]
```

3. El bucle for después del iterador.
```python
[col for col in jug_2b.columns]
```

4. Añadimos el condicional if después del bucle for
```python
[col for col in jug_2b.columns if col not in cols_cat]
```

Si no tenéis claro como llamar al iterador, empezad con el bucle for, pero después acordaros de añadir siempre el iterador al principio de los corchetes.

## 2.2.2 El condicional else dentro de la list comprehension

También podemos utilizar los condicionales if  y else conjuntamente, dentro de la list comprehensions con una pequeña modificación. Ya vimos como el if iba detrás del bucle for pero al añadir tambien un else, estos pasaran al principio de la list comprehension, vemoslo un ejemplo.
