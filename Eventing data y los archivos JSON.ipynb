# 1. Conociendo el eventing data y los archivos JSON

Es hora de adentrarnos en el análisis del eventing data, que son todos los datos recogidos, relacionados con lo que sucede con la pelota. Quién pasa el balón, desde donde lo pasa, qué tipo de pase, donde finaliza el pase, quien recibe el pase, desde donde se tira, en qué acabó el tiro, etc. Toda esta información y muchas más es la que se conoce como eventing data. 

Este tipo de data está guardada en archivos de formato tipo JSON, vimos el archivo tipo CSV que almacenamos la información en un texto plano separada por comas. Con el formato JSON, la cosa cambia, utilizamos diccionarios para almacenar la información. Pasemos a adentrarnos en este apasionante mundo del análisis de datos con el formato JSON. 


```python
import pandas as pd
import json
```

# 1.1. Leer un archivo JSON
Utilizamos la palabra reservada del lenguaje with más la función open() para leer un archivo JSON. De esta manera podremos ver la información que contiene el archivo de datos competitions.json

```python
with open('/content/drive/MyDrive/Mi curso/Data/event_data/Statsbomb/data/competitions.json') as f:
  competitions = json.load(f)
```

# 1.2. Estructura interna de una archivo JSON
Una vez importado el archivo con la función json.load() veamos la estructura interna del archivo JSON

```python
competitions
```

Esto es solo un extracto de la información que contiene, pero como podemos ver no es más que una estructura de diccionarios. Podemos utilizar la función len() para saber la cantidad de diccionarios dentro de nuestro archivo JSON.

```python
len(competitions)
```

Como hemos visto, los diccionarios se encuentran dentro de una lista, por tanto podemos acceder a la información gracias a la técnica de los índices de las listas, que ya hemos ido viendo durante todo el curso. Vamos a Acceder a los valores del primer elemento del archivo JSON.

```python
competitions[0]
```

Ahora con lo aprendido en el curso, podemos ir accediendo a la información, ya que estamos obteniendo un diccionario y nosotros ya sabemos como acceder a la llaves de un diccionario.

```python
competitions[0]['competition_name']
```

Vamos a iterar sobre el valor de todas las llaves de 'competition_name' de cada diccionario.

```python
for competition in competitions:
  print(competition['competition_name'])
```

Creamos una lista vacía, para a través de un bucle for, añadir las diferentes competiciones.  

```python
competition_name = []

for competition in competitions:
  if competition['competition_name'] not in competition_name:
    competition_name.append(competition['competition_name'])

competition_name
```

Realizamos el mismo procedimiento para saber el id de cada competición.

```python
competition_name = []
competition_id = []

for competition in competitions:
  if competition['competition_name'] not in competition_name:
    competition_name.append(competition['competition_name'])

  if competition['competition_id'] not in competition_id:
    competition_id.append(competition['competition_id'])
```

```python
competition_id
```

Utilizamos la función list() y zip() para juntar los valores de los id de las competiciones y el nombre de las competiciones.

```python
list(zip(competition_id, competition_name))
```

Realizamos el mismo procedimiento para extraer el id de la temporada y el año de la temporada.

```python
season_id = []
season_name = []

for season in competitions:
        if season['season_id'] not in season_id:
            season_id.append(season['season_id'])
        if season['season_name'] not in season_name:
            season_name.append(season['season_name'])
```

```python
list(zip(season_id, season_name))
```

Del archivo competitions.json hemos extraído las diferentes competiciones de las que tenemos datos con su respectivo id y también las diferentes temporadas disponibles con su respectivo id.

# 1.4. Extraer la información de los partidos 
Vamos a realizar un bucle for, donde extraemos todos los nombres de los equipos, el resultado y el Id del partido. Para realizar un print y poder ver de una mejor manera los partidos y resultados que tenemos en este archivo JSON.


```python
for match in matches:
  home_team_name = match['home_team']['home_team_name']
  away_team_name = match['away_team']['away_team_name']
  home_score = match['home_score']
  away_score = match['away_score']
  match_id = match['match_id']

  print(home_team_name, 'vs', away_team_name, ':', str(home_score), '-', str(away_score), '|', str(match_id))
```

Ahora tenemos la información de los partidos y de su ID, para entonces poder ir a buscar en la carpeta events los eventos de dicho partido. Vamos a seleccionar el partido del Betis vs Barcelona que terminó en 1-4. Lo que necesito para acceder a los eventos de ese partido es el ID del partido.
